Problem 1: Mobile Recharge Validation System

```python
def validate_recharge():
    valid_plans = [199, 299, 399, 599]

    while True:
        amount = int(input("Enter recharge amount: ₹"))

        if amount < 50:
            print("❌ Minimum recharge amount is ₹50. Try again.")
        elif amount not in valid_plans:
            print("❌ Invalid plan selected. Valid plans are:", valid_plans)
        else:
            print("✅ Recharge successful for ₹", amount)
            break

# Call function
validate_recharge()
```

Problem 2: Inventory Reorder Alert System

```python
def check_inventory(stock_dict):
    for product, quantity in stock_dict.items():
        if quantity < 15:
            print(f"⚠ Reorder Alert: {product} stock is low ({quantity})")
        else:
            print(f"✅ Stock OK: {product} ({quantity})")

# Example dictionary
inventory = {
    "Laptop": 20,
    "Mouse": 10,
    "Keyboard": 5,
    "Monitor": 18
}

check_inventory(inventory)
```

Problem 3: Student Result Processing System

```python
def calculate_result(marks):
    total = 0

    for mark in marks:
        total += mark

    average = total / len(marks)
    print("Average Marks:", average)

    if average >= 50:
        return "Pass"
    else:
        return "Fail"

# Example
student_marks = [60, 70, 55, 40, 65]
result = calculate_result(student_marks)
print("Result:", result)
```

Problem 4: Cab Fare Estimator with Retry Option

```python
def calculate_fare(distance, peak):
    base_fare = 50
    per_km_rate = 12

    fare = base_fare + (distance * per_km_rate)

    if peak == "yes":
        fare += fare * 0.25   # 25% extra

    return fare


while True:
    distance = float(input("Enter distance in km: "))
    peak = input("Is it peak hour? (yes/no): ").lower()

    total_fare = calculate_fare(distance, peak)
    print("Total Fare: ₹", total_fare)

    retry = input("Do you want to calculate again? (yes/no): ").lower()
    if retry != "yes":
        break
```

Problem 5: Employee Attendance Eligibility Checker

```python
def check_eligibility(attendance_list):
    total_days = len(attendance_list)
    present_days = 0

    for day in attendance_list:
        if day == "P":
            present_days += 1

    percentage = (present_days / total_days) * 100
    print("Attendance Percentage:", percentage)

    if percentage >= 75:
        return "Eligible"
    else:
        return "Not Eligible"

# Example
attendance = ["P", "P", "A", "P", "P", "A", "P", "P"]
status = check_eligibility(attendance)
print("Status:", status)
```

Problem 6: Password Strength Checker

```python
def check_password(password):
    has_digit = False
    has_special = False
    special_chars = "@#$"

    if len(password) < 8:
        return "Weak Password (Minimum 8 characters required)"

    for char in password:
        if char.isdigit():
            has_digit = True
        if char in special_chars:
            has_special = True

    if has_digit and has_special:
        return "Strong Password"
    else:
        return "Weak Password (Must include digit and special character)"

# Example
user_password = input("Enter password: ")
print(check_password(user_password))
```
