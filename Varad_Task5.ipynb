Problem 1: Post Engagement Analyzer

```python
def analyze_engagement(likes_list):
    total_likes = 0

    for like in likes_list:
        total_likes += like

    if total_likes >= 1000:
        status = "Viral Post"
    else:
        status = "Normal Engagement"

    print("Total Likes:", total_likes)
    print("Post Status:", status)

# Example
analyze_engagement([200, 300, 150, 400])
```

Problem 2: Medicine Stock Alert

```python
def medicine_stock(stock):

    if stock < 10:
        status = "Low Stock Alert"
    else:
        status = "Stock Sufficient"

    print("Medicine Stock:", stock)
    print("Status:", status)

medicine_stock(6)
```

Problem 3: Rainfall Adequacy Checker

```python
def rainfall_checker(rainfall_data, required_level):
    total = 0

    for rain in rainfall_data:
        total += rain

    average = total / len(rainfall_data)

    if average >= required_level:
        status = "Adequate Rainfall"
    else:
        status = "Inadequate Rainfall"

    print("Average Rainfall:", average)
    print("Rainfall Status:", status)

rainfall_checker([60, 80, 75, 70], 70)
```

Problem 4: Duplicate Account Detection

```python
def detect_duplicates(usernames):

    if len(usernames) != len(set(usernames)):
        result = "Yes"
    else:
        result = "No"

    print("Duplicate Accounts Found:", result)

detect_duplicates(["user1", "user2", "user3", "user1"])
```

Problem 6: Premium Crop Price Filter

```python
def premium_crops(prices):

    premium = []

    for price in prices:
        if price > 2000:
            premium.append(price)

    print("Premium Crops:", premium)

premium_crops([1500, 2500, 3200, 1800])
```

Problem 7: Application Health Checker

```python
def app_health(errors):

    if errors == 0:
        status = "Healthy"
    elif errors <= 5:
        status = "Minor Issues"
    else:
        status = "Critical Issues"

    print("Error Count:", errors)
    print("System Status:", status)

app_health(7)
```

Problem 8: Daily Transaction Checker

Problem 9: Attendance Eligibility

```python
def attendance_checker(attendance_list):

    total = sum(attendance_list)
    percentage = (total / len(attendance_list)) * 100

    if percentage >= 75:
        status = "Eligible"
    else:
        status = "Not Eligible"

    print("Attendance Percentage:", percentage)
    print("Exam Eligibility:", status)

attendance_checker([1,1,1,0,1])
```

Problem 10: Electricity Bill Analyzer

```python
def electricity_bill(units):

    if units <= 100:
        bill = units * 3
    elif units <= 200:
        bill = (100 * 3) + ((units - 100) * 5)
    else:
        bill = (100 * 3) + (100 * 5) + ((units - 200) * 7)

    if bill < 500:
        usage = "Low Usage"
    elif bill <= 1500:
        usage = "Moderate Usage"
    else:
        usage = "High Usage"

    print("Total Bill:", bill)
    print("Usage Status:", usage)

electricity_bill(250)
```
