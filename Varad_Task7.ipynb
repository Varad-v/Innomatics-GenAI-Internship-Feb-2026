Problem 1: Smart Light Controller

```python
class SmartLight:
    def __init__(self, light_name):
        self.light_name = light_name
        self.status = "OFF"

    def turn_on(self):
        self.status = "ON"

    def turn_off(self):
        self.status = "OFF"

    def display_status(self):
        print(f"{self.light_name} is {self.status}")

# User Input
name = input("Enter Light Name: ")
action = input("Enter Action (ON/OFF): ")

light = SmartLight(name)

if action.upper() == "ON":
    light.turn_on()
else:
    light.turn_off()

light.display_status()
```

Problem 2: Employee ID Card System

```python
class Employee:
    def __init__(self, name, emp_id, department):
        self.name = name
        self.emp_id = emp_id
        self.department = department

    def display_id_card(self):
        print("\nEmployee ID Card")
        print("Name:", self.name)
        print("ID:", self.emp_id)
        print("Department:", self.department)

# User Input
name = input("Enter Employee Name: ")
emp_id = input("Enter Employee ID: ")
dept = input("Enter Department: ")

emp = Employee(name, emp_id, dept)
emp.display_id_card()
```

Problem 3: Mobile Contact Record


```python
class Contact:
    def __init__(self, name, phone):
        self.name = name
        self.phone = phone

    def display_contact(self):
        print("\nContact Saved")
        print("Name:", self.name)
        print("Phone:", self.phone)

# User Input
name = input("Enter Contact Name: ")
phone = input("Enter Phone Number: ")

contact = Contact(name, phone)
contact.display_contact()
```

Problem 4: Product Price Tag Generator

```python
class Product:
    def __init__(self, name, price):
        self.name = name
        self.price = price

    def display_price_tag(self):
        print("\nProduct:", self.name)
        print("Price: ₹" + str(self.price))

# User Input
pname = input("Enter Product Name: ")
price = input("Enter Price: ")

product = Product(pname, price)
product.display_price_tag()
```

Problem 5: Movie Rating Display System


```python
class Movie:
    def __init__(self, name, rating):
        self.name = name
        self.rating = rating

    def display_movie(self):
        print("\nMovie:", self.name)
        print("Rating:", self.rating, "/ 5")

# User Input
mname = input("Enter Movie Name: ")
rating = float(input("Enter Rating: "))

movie = Movie(mname, rating)
movie.display_movie()
```

Problem 6: Delivery Address Manager

```python
class Delivery:
    def __init__(self, customer, address):
        self.customer = customer
        self.address = address

    def display_delivery(self):
        print("\nDelivery Details")
        print("Customer:", self.customer)
        print("Address:", self.address)

# User Input
cust = input("Enter Customer Name: ")
addr = input("Enter Address: ")

delivery = Delivery(cust, addr)
delivery.display_delivery()
```
